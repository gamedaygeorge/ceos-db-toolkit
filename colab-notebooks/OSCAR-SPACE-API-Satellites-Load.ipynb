````python
import requests
import pandas as pd

# Base URL for the OSCAR/Space API
base_url = 'https://space.oscar.wmo.int/api/v1/satellites'

# Parameters for the API request
params = {
    'page': 1,  # Adjust the page number if needed
    'sort': 'id',
    'order': 'asc',
    'filterOn': 'orbit',
    'filterValue': 'Sunsynchronous orbit'  # You can adjust the filter value
}

# Making the GET request to the OSCAR/Space API
response = requests.get(base_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    # Display raw data (optional)
    print(data)
else:
    print(f"Failed to retrieve data: {response.status_code}")

# Convert the 'data' part of the response to a pandas DataFrame (if applicable)
# Assuming 'data' is the key in the JSON response that holds the satellite details
df = pd.DataFrame(data)

# Display the DataFrame
df
````

````python
import pandas as pd

# Extracting the embedded satellites data
satellites = data['_embedded']['satellites']

# Creating a DataFrame with only relevant satellite data
# You can customize the columns you'd like to include
df = pd.DataFrame([{
    'ID': sat['id'],
    'Name': sat['fullname'],
    'Acronym': sat['acronym'],
    'Status': sat['status'],
    'Orbit': sat['orbit'],
    'Launch Date': sat['launch_date'],
    'End of Life': sat['EoL'],
    'Space Agency': sat['space_agency']
} for sat in satellites])

# Display the DataFrame
df
````
